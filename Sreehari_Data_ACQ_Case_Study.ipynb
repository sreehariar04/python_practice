```python
import numpy as np
import pandas as pd
```

**Step 1: Load SpaceX Launch Data from API**

```python
import requests  # api request reading
url = " https://api.spacexdata.com/v4/launches"
response = requests.get(url)
launch_data = response.json()
data_df = pd.DataFrame(launch_data)
data_df=data_df[['name', 'date_utc', 'success', 'details', 'rocket']]
data_df
```

```python
data_df['date_utc'] = pd.to_datetime(data_df['date_utc'])
data_df['year'] = data_df['date_utc'].dt.year
data_df
```

**Step 2: Load Rocket Metadata**


```python
import requests  # api request reading
url = "https://api.spacexdata.com/v4/rockets"
response1 = requests.get(url)
rocket_data = response1.json()
rocket_data_df = pd.DataFrame(rocket_data)
rocket_data_df = rocket_data_df[['id', 'name', 'type', 'active', 'stages']]
rocket_data_df
```

**Step 3: Merge Launch and Rocket Data**


```python
merged_df = data_df.merge(
    rocket_data_df,
    left_on='rocket',
    right_on='id',
    how='left',
    suffixes=('_launch', '_rocket')
)
merged_df = merged_df.drop(columns=['id'])
merged_df

```

**Step 4: Add Simulated Country Information**

```python
countries = ['USA', 'Russia', 'India', 'China', 'France']
merged_df['country'] = np.random.choice(countries, size=len(merged_df))
merged_df
```

**Step 5: Store Merged Data in SQLite3**

```python
import sqlite3
conn=sqlite3.connect ('space_df.db')
merged_df.to_sql('launch_data',conn,if_exists='replace',index=False)
conn.commit()
```

**Step 6: Run SQL Queries on the Data to analyze**

**1. Launches by Country**

```python
q1 = """SELECT country ,COUNT(*) AS launch_count
FROM launch_data
GROUP BY country
ORDER BY launch_count DESC;
"""

pd.read_sql(q1, conn)
```

**2. Which year had the highest number of launches?**

```python
q2 = """
SELECT year,COUNT(*) AS launch_count
FROM launch_data
GROUP BY year
ORDER BY launch_count DESC
LIMIT 1;"""
pd.read_sql(q2,conn)
```

**3. Top 5 Missions by Launch Count**

```python
q3 = """
SELECT name_launch ,COUNT(*) AS launch_count
FROM launch_data
GROUP BY name_launch
ORDER BY launch_count DESC
LIMIT 5;"""
pd.read_sql(q3,conn)
```
