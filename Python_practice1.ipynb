2. Create a Python program to calculate the factorial of a given number using a function

```python
#Q2.

def fact(n):# define a function called fact
  f = 1 #initial value of f is 1
  for i in range (1,n+1):
   f = f*i # the initial value if multiplied by numbers in the range in each iteration.
  return f # returns result


x= int(input("Enter a number: "))

print (fact(x))
```

3.Develop a Python program that checks whether a given word or phrase is a palindrome.

```python
#Q3:

p = input("Enter a phrase: ")
rev = p[::-1]

if p==rev:
  print('It is Palindrome')
else:
  print('It is not Palindrome')
```

4.Create a Python program that takes a sentence as input and counts the number of vowels and consonants.

```python
#Q4.

n = input("Enter your statement ")
vowels ='aeiouAEIOU' # initialize vowels

vowel = 0
consonant = 0

for char in n : #checks for character in input
  if char.isalpha():#checks only alphabets
   if char in vowels:# checks for character in vowels
    vowel += 1 #if yes vowels increments
   else:
    consonant+=1#if no consonants is incremented
print('Vowels:',vowel)
print('Consonants:',consonant)
```

7.Develop a Python program that checks if a given year is a leap year.

```python
#Q7.
year=int(input('Enter a year: '))

if year%400==0:
  print('Its a leap year')
elif year%100!=0 and year%4==0:
  print('leap year')
else:
  print('not leap year' )
```

8.Write a Python program that checks if a given number, is a prime number.

```python
#Q8.
def is_prime(n):
 count=0
# a prime number only has 2 divisors .. 1 and the number itself.
 for i in range(1,n+1):
  if(n%i==0):# checks if given number is divisible by each number
    count+=1 # if completely divisible , count is icremented
 if count==2:# checks if count is 2
    return f'{n} is a Prime Number'
 else:
      return f'{n} is not a Prime Number'

n=int(input('enter a number: '))

print(is_prime(n))
```
