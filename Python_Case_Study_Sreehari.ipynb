# Q1. Write a function that return the range (maximum and minimun) of a given list of numbers

```python
def range(number):
  maximum = number[0]
  minimum = number[0]
  for num in number :
    if num>=maximum:
      maximum = num
    elif num<=minimum:
      minimum = num
  print(f'Maximum :',maximum)
  print('Minimum :',minimum)
  print(f'Range is {maximum-minimum}')

list_num = [2,9,2,1,19,15]
range(list_num)
```

## Q2. Write function that returns mean value of a list of numbers

```python
def mean_value(n):
  total=0
  for num in n:
    total+=num
  total_count = len(n)
  mean = total/total_count
  return mean

num=[2,3,5,7,3,5]
print('Mean:', round(mean_value(num),2))
```

# Q6. Write a function that returns the lower bound of a list of numbers

```python
def lower_bound(n):
  minimum_value = n[0]
  for num in n:
    if num < minimum_value:
      minimum_value = num
  return minimum_value

numbers = [34,2,4,45,7,3,1]
lower_bound(numbers)
```

##Q8. Write a function to return standard deviation of a list of numbers

```python
def sd(n):
  mean = sum(n)/len(n)

  squared_diff = 0

  for i in n:
    squared_diff += (i-mean)**2

  sdev = (squared_diff/len(n))**0.5


num= [2,5,7,9]
print (sd(num))
```
