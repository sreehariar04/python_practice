**5) Rename columns ("Title" > "Movie_Title")**

```python
df1 = df.rename(columns={'Title':'Movie_Title'})
df1.head(2)
```

**6) Get unique genres**

```python
unique_genres = (
    df1['Genre']
    .str.split(',')        # split into lists
    .explode()             # turn lists into rows
    .str.strip()           # remove spaces
    .unique()              # get unique values
)
print(unique_genres)
```

**7) Extract number of genres per movie**

```python
df1['Genre_Count'] = df1['Genre'].apply(lambda x: len(x.split(',')))
df1.head(5)
```

8.

**9) Top 5 movies by Rating**

```python
df1.nlargest(5, 'Rating')
```

**10) Movies with highest/lowest revenue**

**Highest Revenue(Millions)**

```python
df1.loc[df1['Revenue (Millions)'].idxmax(), ['Movie_Title', 'Revenue (Millions)']]
```

**Lowest Revenue**

```python
df1.loc[df1['Revenue (Millions)'].idxmin(), ['Movie_Title', 'Revenue (Millions)']]
```

**11) Avg. rating by Director**

```python
df1.groupby('Director')['Rating'].mean()
```

**12) Movies per year**

```python
movies_per_year = df1['Year'].value_counts()
print(movies_per_year)
```

**13) Movies with rating > 8 & revenue > 100M**

```python
df1[(df1['Rating'] > 8) & (df1['Revenue (Millions)'] > 100)][["Movie_Title","Rating","Revenue (Millions)"]]
```

14) **Group by Director: avg Rating, Revenue**

```python
df1.groupby('Director')[['Rating', 'Revenue (Millions)']].mean()
```

**15) Group by Year: count, avg Rating**

```python
year_status = df1.groupby("Year").agg(
    movie_count=("Movie_Title", "count"),
    avg_rating=("Rating", "mean")
)

year_status
```

16) Create a new column Profitability:
a) Revenue divided by Duration
b) Find top 10 most profitable movies

```python
df1['Profitability'] = df1['Revenue (Millions)'] / df1['Runtime (Minutes)']
top_10=df1.nlargest(10,"Profitability")
top_10[["Movie_Title","Revenue (Millions)","Runtime (Minutes)","Profitability"]]
```

17) Create a column that categorizes movies by rating:
a) Rating < 6: Poor
b) 6 â‰¤ Rating < 7.5: Average
c) Rating â‰¥ 7.5: Excellent


```python
df1["Rating_Summary"]=df1["Rating"].apply(lambda x: 'Poor' if x < 6 else ('Average' if x < 7.5 else 'Excellent'))
df1[['Movie_Title', 'Rating', 'Rating_Summary']]
```

18) Create a custom function to score movies based on:
i) Apply row['Rating'] * 10 + row['Revenue (Millions)'] as High score
Apply a function to dataset to assign the score

```python
df1['High_Score'] = df1.apply(lambda row: row['Rating'] * 10 + row['Revenue (Millions)'], axis=1)
df1[['Movie_Title', 'Rating', 'Revenue (Millions)', 'High_Score']]
```
